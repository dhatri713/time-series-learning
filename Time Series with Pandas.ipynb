## Resampling data based of time

```python
#resample this daily data to yearly data
df.resample(rule='A')
#basically we want to take all those daily datapoints and aggregate them
# somehow so that they become yearly datapoints
#we also need to mention how we're aggregating it
df.resample(rule='A').mean()
```

```python
# we're not limited to pandas aggregating fumctions.
#we can make our own
```

```python
def first_day(entry):
    # is there an entry
    if len(entry) != 0:
        # if so, return first entry
        return entry[0]
```

```python
df.resample(rule='A').apply(first_day)
# returns the first day's closing price and value of stock traded
```

```python
df.head()
```

```python
only_show_close_resampled = df['Close'].resample('A').mean()
```

```python
%matplotlib inline
only_show_close_resampled.plot.bar()
```

```python
from matplotlib import dates
```

```python
ax = df['Close'].plot(xlim=['2017-01-01', '2017-03-01'], ylim=[50,60], figsize=(12, 5))
ax.set(xlabel='')

ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter('%d'))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n%b'))

ax.yaxis.grid(True)
ax.xaxis.grid(True)
```
